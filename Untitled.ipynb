```python
import pandas as pd
import numpy as np
from kerasOOP import keras_ann, ann_data
import os
from os import listdir
import json
import sys
```

```python
myloc = os.path.expanduser('~') + "/Documents/201912/"
print(myloc)
```

```python
dataLoc = myloc + 'eegData/'
loadLoc = myloc + 'kerasTimeSeries/myweights/'
```

```python
myAnn = keras_ann()
myData = ann_data(dataPath=dataLoc)
```

## Load Weight Locations and Model Lists

```python
allWeights = []
models = []
freqs = ['delta','theta','alpha','beta1','beta2']

# there are several models to choose from,
# manually choose the index of the model 
modelChoice = {}
for freq in freqs:
    modelChoice[freq] = 0
modelChoice['delta'] = 1

for freq in freqs:
    weights = []
    myAnn.getWeights(weights,loadLoc + freq)
    allWeights.append(loadLoc + freq + '/' + weights[modelChoice[freq]][0])
    #print(weights)
    #print(weights[modelChoice[freq]][0])
    for filename in listdir(loadLoc + freq):
        if 'Model' in filename:
            with open(loadLoc + freq + '/' + filename) as tfile:
                modelIdtemp = 0
                while modelIdtemp < modelChoice[freq]:
                    tfile.readline()
                    modelIdtemp += 1
                #print(f"model: {modelIdtemp}")
                models.append(json.loads(tfile.readline().split('|')[-1].strip()))
```

```python
print(models[0])
```

```python
print(allWeights)
```

## Get Conv Weights for Each Model
- Load Data
- Load Model
- Load Weights into Model
- Run Model
- get Weights

```python
modelNum = 0
convWeights = []
for paramSet in models:
    freq = freqs[modelNum]
    try:
        print("loadingData")
        myData.readData() 
        [normSTD, normMean] = myAnn.getNorm(loadLoc + freq + '/')
        myData.expandDims()
        myData.normalize(normSTD=normSTD, normMean=normMean)
        print("ERROR3")
        weightSet = allWeights[modelNum]
        print(weightSet)
        print(loadLoc)
        print("printModel")
        convWeights.append(myAnn.printModel(paramSet, weightSet, X=myData.data))
        
    except Exception as e:
        print("ERROR",sys.exc_info()[0])
    modelNum += 1
```

```python
print(convWeights[1][0])
```

```python
convFilters = []
for convWeight in convWeights:
    convFilters.append([])
    shape = convWeight[0].shape[0]
    count = convWeight[0].shape[2]
    for c in range(count):
        convFilters[-1].append([])
        for s in range(shape):
            convFilters[-1][-1].append(convWeight[0][s,0,c])
```

```python
for filts in convFilters:
    print("="*11)
    print()
    for filt in filts:
        print(filt)
```
